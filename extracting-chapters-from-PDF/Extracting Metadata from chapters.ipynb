```python
def get_last_PDF(pdf_list):
    pdf_list = pdf_list.replace("[","")
    pdf_list = pdf_list.replace("]","")
    pdf_list = pdf_list.replace("'","")
    pdf_list = pdf_list.replace('"','')
    pdf_list = pdf_list.split(",")
    a = pdf_list[len(pdf_list) - 1]
    a.strip()
    return(a)


def get_last_page(page_list_str):
    if page_list_str == '[]':
        return(0)
    
    page_lists = page_list_str.split("]],[[")
    page_list_2D = []
    
    for page_list in page_lists:
        page_list_2D.append(page_list.split(","))
        
    llist = []
    for i in range(len(page_list_2D)):
        l = []
        for j in range(len(page_list_2D[i])):
            a = page_list_2D[i][j]
            a = a.replace("[","")
            a = a.replace("]","")
            a = a.replace("'","")
            a = a.replace('"','')
            a.strip
            l.append(a)
        llist.append(l)
    
    result = llist[len(llist) - 1][len(llist[len(llist)-1])-1]
                                
    return(int(result))
```

```python
Key_Found = []
PDF_Found_Last = []
Page_Found_Last = []

for i in range(len(allChaptersDF)):
    PDF_Found_Last_ele = get_last_PDF(allChaptersDF.iloc[i]['PDF_Found'])
    PDF_Found_Last.append(PDF_Found_Last_ele)
    Page_Found_Last_ele = get_last_page(allChaptersDF.iloc[i]['Page_Found'])
    Page_Found_Last.append(Page_Found_Last_ele)
    Key_Found.append(PDF_Found_Last_ele + ' - ' + str(Page_Found_Last_ele))
    
#allChaptersDF['Key_Found'] = Key_Found
allChaptersDF['Page_Found_Last'] = Page_Found_Last
allChaptersDF['PDF_Found_Last'] = PDF_Found_Last
allChaptersDF
```

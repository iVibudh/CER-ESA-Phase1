```python
import pandas as pd
from PyPDF2 import PdfFileReader
```

```python
def get_info(path):
    with open(path, 'rb') as f:
        pdf = PdfFileReader(f)
        info = pdf.getDocumentInfo()
        number_of_pages = pdf.getNumPages()
 
    print(info)
 
    author.extend([info.author])
    creator.extend([info.creator]) 
    producer.extend([info.producer])
    subject.extend([info.subject])
    title.extend([info.title])
```

```python
df_outlines = pd.read_csv("F:\Environmental Baseline Data\Version 3\Indices\Outline_Present_htmlTOC.csv")
 
len(df_outlines)
df_outlines[20:30].head(5)
```

```python
author = []
creator = []
producer = []
subject = []
title = []

for path in df_outlines['file_paths']:
    try:
        get_info(path)
    except:
        author.extend(['NA'])
        creator.extend(['NA']) 
        producer.extend(['NA'])
        subject.extend(['NA'])
        title.extend(['NA'])
    
```

```python
df_outlines['Author'] = author
df_outlines['Creator'] = creator 
df_outlines['Producer'] = producer
df_outlines['Subject'] = subject
df_outlines['Title'] = title


df_outlines.head(5)
```

```python
df_outlines.to_csv("F:\Environmental Baseline Data\Version 3\Indices\Outline_Present_withNotSoUsefulMetahtmlTOC.csv")
```

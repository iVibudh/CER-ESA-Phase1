### Load csv's as dataframes and check if the dataframe is empty, if yes, remove the empty csv from the directory

```python
for x in os.listdir(r'F:\Environmental Baseline Data\Version 4 - Final\CSV2'):
    cnt = 0
    print(x)
    csv_list = glob.glob(r'\\luxor\data\Branch\Environmental Baseline Data\Version 4 - Final\CSV2\{}\*.csv'.format(x))
    csv_list = [csv_list[i].split('\\')[-1] for i in range (len(csv_list))]
    all_df = list()
    csv_flist = list()
    for i in csv_list:
        path = r'F:\Environmental Baseline Data\Version 4 - Final\CSV2\{}\{}'.format(x,i)
        df_content = list()
        try:
            df = pd.read_csv(path , engine='python')
            col_name = df.columns
            for cn in col_name:
                lst = df[cn].dropna().tolist()
                df_content.extend(lst)
            if len(df_content) == 0:
                os.remove(path)
            else:
                all_df.append(df_content)
        except:
            cnt = cnt + 1 
            print(path)
            print(cnt)
    
```
